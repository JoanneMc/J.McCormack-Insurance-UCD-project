```python
#Begin merge with Employee_data & insurance data using common heading 'Agent_ID' called Claims

Claims = pd.merge(Employee_data,Insurance_data,how = 'left',on = 'AGENT_ID')
```

```python
#Complete merge using Claims and Vendor_data - new Dataframe created called Total_Claims_df link data using vendor_ID

Total_Claims_df = pd.merge(Claims,Vendor_data, how = 'left', on = 'VENDOR_ID')
```

```python
# We know a large percentage of Address Line 2 for all data is blank, therefore removing same for all address line 2

Total_Claims_df.drop(['ADDRESS_LINE2_x','ADDRESS_LINE2_y','ADDRESS_LINE2'], axis = 1)
```

```python
#I want to change the address column labels to be associated with Agent, customer & vendor to make ot easier to distinguish same
# EG Changing names to Agent address rathe than address line

Total_Claims_df.rename(columns = {'ADDRESS_LINE1_x':'AGENT ADDRESS', 'CITY_x':'AGENT CITY', 'STATE_x':'AGENT STATE','POSTAL_CODE_x':'AGENT POSTAL CODE', 'ADDRESS_LINE1_y':'CUSTOMER ADDRESS','CITY_y':'CUSTOMER CITY','STATE_y':'CUSTOMER STATE','POSTAL_CODE_y':'CUSTOMER POSTAL CODE', 'ADDRESS_LINE1':'VENDOR ADDRESS','CITY':'VENDOR CITY', 'STATE':'VENDOR STATE','POSTAL CODE':'VENDOR POSTALCODE'},inplace=True)
```

```python
# I wanted to see what the most expensive insurance type is by sorting the highest premium amount which equates to health 

TotalPremium = Total_Claims_df.sort_values("PREMIUM_AMOUNT",ascending = False)
TotalPremium[['PREMIUM_AMOUNT','INSURANCE_TYPE']]
```

```python
#I want to see the breakdown of premium per family member in each household, I want to add this as a new column in the DataFrame. 

def Price_householder(Total_Claims_df):
    Price_Household = Total_Claims_df['PREMIUM_AMOUNT'] / Total_Claims_df['NO_OF_FAMILY_MEMBERS']
    return round(Price_Household,2)
    
```

```python
Total_Claims_df.apply(Price_householder, axis=1)
```

```python
Total_Claims_df['Price_householder'] = Total_Claims_df.apply(Price_householder, axis =1)
Total_Claims_df
```

```python
#The table below demonstrates that yes health is the highest in premium revenue but certainly not the highest in claim amount.
#Claim amount is certainly highest under life insurance followed by property claims. 

plt.figure(figsize=(15,8))
sns.scatterplot(x='CLAIM_AMOUNT',y='PREMIUM_AMOUNT', data=Total_Claims_df, hue='INSURANCE_TYPE').set_title('Claim vs Premium')
```

```python
#I wanted to count the total amount paid against each insurance product

Total_Claim_Amount = pd.DataFrame(Total_Claims_df.groupby(['INSURANCE_TYPE'])['CLAIM_AMOUNT'].sum())
Total_Claim_Amount['INSURANCE_TYPE'] = Total_Claim_Amount.index
Total_Claim_Amount.reset_index(drop= True, inplace= True)
Total_Claim_Amount
```

```python
#I wanted to visualize the value range of each claim payment per insurance product

plt.figure(figsize=(14,11))
sns.boxplot(data=Total_Claims_df, x ="INSURANCE_TYPE", y="CLAIM_AMOUNT",).set(title='Claim per product')
```

```python
#Similar to the claim amount, I wanted to replicate that function to guage the total premium brought in under each product.

Total_Premium_Amount = pd.DataFrame(Total_Claims_df.groupby(['INSURANCE_TYPE'])['PREMIUM_AMOUNT'].sum())
Total_Premium_Amount['INSURANCE_TYPE'] = Total_Premium_Amount.index
Total_Premium_Amount.reset_index(drop= True, inplace= True)
Total_Premium_Amount
```

```python
# Visualizing the range of premium price per product

plt.figure(figsize=(14,11))
sns.boxplot(data=Total_Claims_df, x ="INSURANCE_TYPE", y="PREMIUM_AMOUNT",).set(title='Premium per product')
```
